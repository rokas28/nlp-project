```python
import spacy
from spacy import displacy
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# EN
en = spacy.load("en_core_web_sm")
doc = open('Pradinis_en.txt', encoding="utf8").read()
sents = en(doc)
people = [ee for ee in sents.ents if ee.label_ == 'PERSON']
text=sents.text
for a in people:
    text=text.replace(a.text, 'ANONIMAS')
myfile = open('Galutinis_en.txt', 'w', encoding="utf8")
myfile.write(text)
myfile.close()

# LT
lt = spacy.load("lt_core_news_sm")
doc = open('Pradinis_lt.txt', encoding="utf8").read()
sents = lt(doc)
people_lt = [ee for ee in sents.ents if ee.label_ == 'PERSON']
text=sents.text
for a in people_lt:
    text=text.replace(a.text, 'ANONIMAS')
myfile_lt = open('Galutinis_lt.txt', 'w', encoding="utf8")
myfile_lt.write(text)
myfile_lt.close()

# Plots
labels = ['EN', 'LT']
right = [64, 31]
missed = [8, 20]
wrong = [3, 4]

x = np.arange(len(labels))  
width = 0.3  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, right, width, label='Teisingai pakeisti vardai')
rects2 = ax.bar(x, wrong, width, label='Neteisingai pakeisti vardai')
rects3 = ax.bar(x + width, missed, width, label='Nepakeisti vardai')

ax.set_ylabel('Vardų skaičius')
ax.set_title('Tekste esančių vardų pakeitimas žodžiu ANONIMAS')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)

fig.tight_layout()

fig.set_figheight(6)

plt.show()

#---------------------------------------------
labels = ['EN', 'LT']

a = round(64*100/72, 2)
b = round((31*100/55), 2)
precision = [float(a), float(b)]

x = np.arange(len(labels))  
margin = 0.05
width = (1.-2.*margin)  

fig, ax = plt.subplots()
rects1 = ax.bar(x, precision, width, label='Vardų keitimo tiksumas')

ax.set_ylabel('Tikslumas %')
ax.set_title('Tekste esančių vardų pakeitimo žodžiu ANONIMAS tikslumas')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)

fig.tight_layout()

fig.set_figheight(6)

plt.show()
```
